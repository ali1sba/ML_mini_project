```python
HealthTrain = HealthTrain.drop("case_id", axis=1)
HealthTest = HealthTest.drop("case_id", axis=1)
```

```python
XTrain = HealthTrain.iloc [:,:-1]
yTrain = HealthTrain.iloc[:,-1]
```

```python
import math
def euclidean_distance(row1, row2):
    distance=0
    for i in range(len(row1)):
        if i in [4,8,13,15]:
            distance += (row1[i] - row2[i])**2
        else:
            distance+=1-(row1[i] == row2[i])
    return math.sqrt(distance)
```

```python
HealthTest=HealthTest.transpose()
XTrain=XTrain.transpose()
```

```python
distances=np.empty((len(HealthTest.transpose()),len(XTrain.transpose())))
print(len(XTrain.transpose()),len(HealthTest.transpose()))
```

```python
for rowTest in range(len(HealthTest.transpose())):
    for rowTrain in range(len(XTrain.transpose())):
        print("row",rowTest ," with row",rowTrain)
        distances[rowTest][rowTrain]=euclidean_distance(HealthTest[rowTest], XTrain[rowTrain])
        print(distances[rowTest][rowTrain])
print(distances)
```

```python
distances1=np.empty((5,7))
for rowTest in range(5):
    for rowTrain in range(7):
        print("row",rowTest ," with row",rowTrain)
        distances1[rowTest][rowTrain]=euclidean_distance(HealthTest[rowTest], XTrain[rowTrain])
        print(distances1[rowTest][rowTrain])
print(distances1)
```

```python
from collections import Counter
yTest=np.empty((5,1), dtype='S')

print(len(distances1))
k=5
for i in range(len(distances1)):
  idx = np.argpartition(distances1[i], k)
  print(yTrain[idx[:k]])
  c = Counter(yTrain[idx[:k]])
  yTest[i][0]=c.most_common(1)[0][0]
  print(c.most_common(1)[0])
print(yTest)

```
