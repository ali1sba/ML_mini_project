```python
HealthTrain = HealthTrain.drop("case_id", axis=1)
HealthTest = HealthTest.drop("case_id", axis=1)
```

```python
XTrain = HealthTrain.iloc [:,:-1]
yTrain = HealthTrain.iloc[:,-1]
XTest = HealthTrain.iloc [:,:-1]
yTest = HealthTrain.iloc[:,-1]
```

```python
yTrain.head()
```

```python
import math
def euclidean_distance(row1, row2):
    distance=0
    for i in range(len(row1)):
        if i in [4,8,13,15]:
            distance += (row1[i] - row2[i])**2
        else:
            distance+=1-(row1[i] == row2[i])
    return math.sqrt(distance)
```

```python
XTest=XTest.transpose()
XTrain=XTrain.transpose()
```

```python
distances=np.empty((len(XTest.transpose()),len(XTrain.transpose())))
print(len(XTrain.transpose()),len(XTest.transpose()))
```

```python
for rowTest in range(len(XTest.transpose())):
    for rowTrain in range(len(XTrain.transpose())):
        print("row",rowTest ," with row",rowTrain)
        distances[rowTest][rowTrain]=euclidean_distance(XTest[rowTest], XTrain[rowTrain])
        print(distances[rowTest][rowTrain])
print(distances)
```

```python
lenTest=len(XTest.transpose())
```

```python
yTest2=np.empty((lenTest,1), dtype='U100')
k=5
print(len(distances))
for i in range(len(distances)):
  idx = np.argpartition(distances[i], k)
  print(idx)
  print(yTrain[idx[:k]])
  c = Counter(yTrain[idx[:k]])
  yTest2[i][0]=c.most_common(1)[0][0]
  print(c.most_common(1)[0])
print(yTest2)
```

```python
sum =0
for i in range(len(yTest)):
   sum+=yTest[i] == yTest2[i]
print(sum,len(yTest))
accuracy=sum/len(yTest)
print(str(accuracy*100)+'%')
```
