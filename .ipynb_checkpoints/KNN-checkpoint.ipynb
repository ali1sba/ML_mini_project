```python
import math
def euclidean_distance(row1, row2):
    distance=0
    print("row1-----row2-----")
    for i in range(len(row1)):
        if i in [4,8,13,15]:
            distance += (row1[i] - row2[i])**2
        else:
            distance+=1-(row1[i] == row2[i])
    return math.sqrt(distance)
```
